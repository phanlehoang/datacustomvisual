```python
import json
import pandas as pd
# Đọc dữ liệu từ file JSON
with open('sample json.txt') as f:
    data = json.load(f)

# In ra dữ liệu đã đọc
print(data)
```

```python
# In data theo từng cặp key-value
for key, value in data.items():
    print(key, value)
```

```python
data = json.dumps(data)
data = json.loads(data)
data
```

```python
# Phân tích cấu trúc của chuỗi JSON, thiết kế mô hình DB phù hợp tạo relationship giữa các bảng. Ví dụ tách thành các bảng như sau:
# Bảng workspaces (danh sách các workspaces)
# Bảng items (danh sách các resources trong workspaces)
# Bảng users (danh sách các users và quyền sử dụng items tương ứng)


# Tạo bảng workspaces
workspaces = pd.DataFrame(data['workspaces'])
workspaces

```

```python
# Chuyển thành file csv
workspaces.to_csv('workspaces.csv', index=False, encoding='utf-8-sig')

# lưu file csv vào folder

```

```python

# Lấy danh sách người dùng từ bộ dữ liệu
users = []
workspaces = data["workspaces"]
for workspace in workspaces:
    kustodatabases = workspace["kustodatabase"]
    for kustodatabase in kustodatabases:
        user_list = kustodatabase["users"]
        for user in user_list:
            user_info = {
                "workspace_id": workspace["id"],
                "kustodatabase_id": kustodatabase["id"],
                "user_id": user["userId"],
                "access_right": user["artifactUserAccessRight"]
            }
            users.append(user_info)

# Xây dựng DataFrame từ danh sách người dùng
user_df = pd.DataFrame(users)

# Hiển thị bảng user
print(user_df)
```
