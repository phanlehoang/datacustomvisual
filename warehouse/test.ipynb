```python
import json
import pandas as pd
# c d liu t file JSON
with open('sample json.txt') as f:
    data = json.load(f)

# In ra d liu  c
print(data)
```

```python
# In data theo tng cp key-value
for key, value in data.items():
    print(key, value)
```

```python
data = json.dumps(data)
data = json.loads(data)
data
```

```python
# Phn tch cu trc ca chui JSON, thit k m hnh DB ph hp to relationship gia cc bng. V d tch thnh cc bng nh sau:
# Bng workspaces (danh sch cc workspaces)
# Bng items (danh sch cc resources trong workspaces)
# Bng users (danh sch cc users v quyn s dng items tng ng)


# To bng workspaces
workspaces = pd.DataFrame(data['workspaces'])
workspaces

```

```python
# Chuyn thnh file csv
workspaces.to_csv('workspaces.csv', index=False, encoding='utf-8-sig')

# lu file csv vo folder

```

```python

# Ly danh sch ngi dng t b d liu
users = []
workspaces = data["workspaces"]
for workspace in workspaces:
    kustodatabases = workspace["kustodatabase"]
    for kustodatabase in kustodatabases:
        user_list = kustodatabase["users"]
        for user in user_list:
            user_info = {
                "workspace_id": workspace["id"],
                "kustodatabase_id": kustodatabase["id"],
                "user_id": user["userId"],
                "access_right": user["artifactUserAccessRight"]
            }
            users.append(user_info)

# Xy dng DataFrame t danh sch ngi dng
user_df = pd.DataFrame(users)

# Hin th bng user
print(user_df)
```
